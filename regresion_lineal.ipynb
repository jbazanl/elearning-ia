```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de entrada
X = np.array([
    [-1, -1],
    [-1, 1],
    [0, 2],
    [-4, 2],
    [-4, 4],
    [-5, 1]
])

# Etiquetas
y = np.array([1, 1, 1, -1, -1, -1])

# Inicializar los pesos y el sesgo
w = np.zeros(X.shape[1])
b = 0

# Función de predicción
def predict(x):
    return np.sign(np.dot(x, w) + b)

# Algoritmo de aprendizaje del perceptrón
def perceptron(X, y, learning_rate=1.0, max_iter=1000):
    global w, b
    for _ in range(max_iter):
        for i in range(len(y)):
            if y[i] * predict(X[i]) <= 0:
                w += learning_rate * y[i] * X[i]
                b += learning_rate * y[i]

# Entrenar el modelo
perceptron(X, y)

# Mostrar los parámetros finales
print("Pesos:", w)
print("Sesgo:", b)

# Calcular los puntos de la frontera de decisión
# x2 = -(w[0] * x1 + b) / w[1]
x1_vals = np.linspace(min(X[:, 0]), max(X[:, 0]), 100)
x2_vals = -(w[0] * x1_vals + b) / w[1]

# Graficar los puntos de datos y la frontera de decisión
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Clase +1')
plt.scatter(X[y == -1][:, 0], X[y == -1][:, 1], color='red', label='Clase -1')
plt.plot(x1_vals, x2_vals, color='black', label='Frontera de decisión')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de entrada
X = np.array([
    [-1, -1],
    [-1, 1],
    [0, 2],
    [-4, 2],
    [-4, 4],
    [-5, 1]
])

# Etiquetas
y = np.array([1, 1, 1, -1, -1, -1])

# Inicializar los pesos y el sesgo
w = np.zeros(X.shape[1])
b = 0

# Algoritmo de aprendizaje del perceptrón
def perceptron(X, y, learning_rate=1.0, max_iter=1000):
    global w, b
    for epoch in range(max_iter):
        errors = 0
        for i in range(len(y)):
            if y[i] * (np.dot(X[i], w) + b) <= 0:
                w += learning_rate * y[i] * X[i]
                b += learning_rate * y[i]
                errors += 1
        # Verificar si no hubo errores en esta iteración
        if errors == 0:
            print(f"Primera predicción exitosa en la época {epoch + 1}")
            print("Pesos (theta):", w)
            print("Sesgo (theta_0):", b)
            return w, b
    print("No se encontró una predicción exitosa en el número máximo de iteraciones.")
    return w, b

# Entrenar el modelo y obtener los parámetros
w, b = perceptron(X, y)

# Calcular los puntos de la frontera de decisión
x1_vals = np.linspace(-6, 1, 100)
x2_vals = -(w[0] * x1_vals + b) / w[1]

# Graficar los puntos de datos y la frontera de decisión
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Clase +1')
plt.scatter(X[y == -1][:, 0], X[y == -1][:, 1], color='red', label='Clase -1')
plt.plot(x1_vals, x2_vals, color='black', label='Frontera de decisión')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Datos de entrada
X = np.array([
    [-1, -1],
    [-1, 1],
    [0, 2],
    [-4, 2],
    [-4, 4],
    [-5, 1]
])

# Etiquetas
y = np.array([1, 1, 1, -1, -1, -1])

# Crear y entrenar el modelo SVM con kernel lineal
clf = SVC(kernel='linear', C=1e10)  # C grande para minimizar el margen
clf.fit(X, y)

# Obtener los parámetros de la frontera de decisión
w = clf.coef_[0]
b = clf.intercept_[0]

print("Pesos (theta):", w)
print("Sesgo (theta_0):", b)

# Calcular los puntos de la frontera de decisión
x1_vals = np.linspace(-6, 2, 100)
x2_vals = -(w[0] * x1_vals + b) / w[1]

# Graficar los puntos de datos y la frontera de decisión
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='blue', label='Clase +1')
plt.scatter(X[y == -1][:, 0], X[y == -1][:, 1], color='red', label='Clase -1')
plt.plot(x1_vals, x2_vals, color='black', label='Frontera de decisión')

# Graficar los vectores de soporte
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1],
            s=100, facecolors='none', edgecolors='k', label='Vectores de soporte')

plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```
